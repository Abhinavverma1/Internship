```python
product_urls=[]
for  i in range(0,3):
    for i in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-4']/a"):
        product_urls.append(i.get_attribute("href"))
    nxt_button=driver.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]').get_attribute('href')#scraping the list of buttons from the page
    driver.get(nxt_button)
```

```python
len(product_urls)
```

```python
product_urls
```

```python
product_name=[]

for i in product_urls:
    driver.get(i)
    try:
        pname= driver.find_element_by_xpath('//span[@class="a-size-large product-title-word-break"]')
        product_name.append(pname.text.replace("/n","new line"))
    except:
        product_name.append('-')
```

```python
len(product_name)
```

```python
brand=[]
price=[]
reurn=[]
Expected_delivery =[]
avail=[]
product_url =[]

for i in product_urls:
    driver.get(i)
    try:
        avl= driver.find_element_by_xpath('//span[@class="a-size-medium a-color-success"]')
        avail.append(avl.text)
    except:
        avail.append('-')
    try:
        p_brand= driver.find_element_by_xpath('//table[@class="a-size-base a-text-bold"]')
        brand.append(p_brand.text)
    except:
        brand.append('-')
     #fetching price
    try:
        pprice= driver.find_element_by_xpath('//span[@class="a-price a-text-price a-size-medium apexPriceToPay"]')
        price.append(pprice.text)
    except:
        price.append('-')
        
        #fetching expected delivery
    try:
        ed= driver.find_element_by_xpath('//div[@id="ddmDeliveryMessage"]')
        Expected_delivery.append(ed.text)
    except:
        Expected_delivery.append('-')
        
                 #fetching return
    try:
        rt= driver.find_element_by_xpath('//div[@id="RETURNS_POLICY"]')
        reurn.append(rt.text)
    except:
        reurn.append('-')
```

```python
len(brand),len(price),len(reurn),len(Expected_delivery),len(avail)
```
